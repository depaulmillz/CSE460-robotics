```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2 * np.pi, 100)

def elipse(a, b, t):
    return np.array([a * np.cos(t), b * np.sin(t)])

def elipse_shifted(a, b, t, init):
    return elipse(a, b, t) + np.array([init[0] - a, init[1]])


def rot(theta):
    return np.array([[-np.cos(theta), np.sin(theta)],[-np.sin(theta), -np.cos(theta)]])
```

```python
duck = np.array([-1.8, -0.3])
init = np.array([1.0, -0.3])

dist = np.linalg.norm(duck - init)

points = np.array([elipse_shifted(dist / 2, 0.65, x, init) for x in t])

plt.plot(points[:,0],points[:,1])

x = [duck[0]]
y = [duck[1]]

plt.plot(x, y, marker = 'x', markersize = 10)
plt.arrow(points[0,0], points[0, 1], points[1,0]-points[0,0], points[1,1]-points[0,1], shape='full', lw=10, length_includes_head=True)
plt.show()
```

```python
duck = np.array([-1.8, -0.3])
init = np.array([0.11, 1.3])

err = duck - init

dist = np.linalg.norm(err)

print(err)

angle = np.arctan2(err[1], err[0])

print(angle)

print(rot(angle))

duck_rot = np.matmul(rot(-angle), duck)
init_rot = np.matmul(rot(-angle), init)

print("Init rot", init_rot)
print("Duck rot", duck_rot)

points = np.array([np.matmul(rot(angle), elipse_shifted(dist / 2, 0.65, x, init_rot)) for x in t])

plt.plot(points[:,0],points[:,1])

x = [duck[0]]
y = [duck[1]]

plt.plot(x, y, marker = 'x', markersize = 10)
plt.arrow(points[0,0], points[0, 1], points[1,0]-points[0,0], points[1,1]-points[0,1], shape='full', lw=10, length_includes_head=True)
plt.show()
```
