```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2 * np.pi, 100)

def elipse(a, b, t):
    return np.array([-b * np.cos(t), a * np.sin(t)])
```

```python
points = np.array([elipse(2, 1, x) for x in t])

plt.plot(points[:,0],points[:,1])
plt.arrow(points[0,0], points[0, 1], points[1,0]-points[0,0], points[1,1]-points[0,1], shape='full', lw=10, length_includes_head=True)
plt.show()
```
