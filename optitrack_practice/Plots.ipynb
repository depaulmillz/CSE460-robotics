```python
part2 = pd.read_csv('part2res.csv')
part2['angle'] = (((part2['angle'] + np.pi) % (2 * np.pi)) - np.pi) / np.pi * 180
part3_q2_q1 = pd.read_csv('part3_q2_q1_res.csv')
part3_q3 = pd.read_csv('part3_q3_results.csv')
part3_q4 = pd.read_csv('spline.csv')
```

```python
fig = plt.figure()
ax = plt.axes()
ax.set_xlabel("Time (s)")
ax.set_ylabel("Angle Difference")
plt.grid()
lineA = ax.plot(part2['time'],part2['angle'], '-g', label="Angle")
ax2 = ax.twinx()
lineB = ax2.plot(part2['time'],part2['dist'], '-b', label="Distance")
ax2.set_ylabel("Distance")

lines = lineA + lineB

labels = [l.get_label() for l in lines]

plt.legend(lines, labels)

plt.savefig("angleAndDist.png")
```
